```python
import joblib
import pandas as pd

# Load trained model
today_model = joblib.load("../models/tesla_price_model.pkl")
future_model = joblib.load("../models/future_price_model.pkl")

# Raw input data (order doesn't matter here)
data = {
    "Open": 340,
    "High": 345,
    "Low": 338,
    "Volume": 85000000,
    "Daily_Change": 2.5,
    "High_Low_Diff": 7,
    "MA_5": 342,
    "MA_10": 339,
    "Pct_Change": 0.7
}

# Convert to DataFrame and reorder correctly
df_input = pd.DataFrame([data])
# Reorder columns to match training data
df_input = df_input[today_model.feature_names_in_]

# Predict today's closing price
today_pred = today_model.predict(df_input)[0]

# Predict Tomorrow's closing price
future_pred = future_model.predict(df_input)[0]


print("Predicted Today Close:", today_pred)
print("Predicted Tomorrow Close:", future_pred)

```
